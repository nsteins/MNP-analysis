```python
X = pd.read_table('~/Documents/Data/5-12-16/Nabond ND on CPW 10mW -1to1 scan 1 x102 y79 1A T0.txt').values
Y = pd.read_table('~/Documents/Data/5-12-16/Nabond ND on CPW 10mW -1to1 scan 1 x102 y79 1A T90.txt').values
Z = pd.read_table('~/Documents/Data/5-12-16/Nabond ND on CPW 10mW -1to1 scan 1 x102 y79 1A T0 X14.txt').values
p = figure(tools=TOOLS)
p.line(X[:,0],X[:,1],color="red")
p.line(Y[:,0],Y[:,1],color="green")
p.line(Z[:,0],Z[:,1],color="blue")
show(p)
```

```python
def lorentzian(X,x0,c):
    return c**2/(np.pi*c*(c**2+(X-x0)**2))
```

```python
def ODMRfit(X,a1,f1,w1,a2,f2,w2):
    return a1*lorentzian(X,f1,w1)+a2*lorentzian(X,f2,w2)
```

```python
from scipy.optimize import curve_fit
popt, pcov = curve_fit(ODMRfit,X[:,0],X[:,1],p0=(-4E8,2.7E9,1E7,-4E8,3E9,1E7))
```

```python
popt
```

```python
p = figure(tools=TOOLS)
poptX, pcov = curve_fit(ODMRfit,X[:,0],X[:,1],p0=(-4E8,2.7E9,1E7,-4E8,3E9,1E7))
p.circle(X[:,0],X[:,1],color="red")
p.line(X[:,0],ODMRfit(X[:,0],*poptX),color='maroon')

poptY, pcov = curve_fit(ODMRfit,Y[:,0],Y[:,1],p0=(-4E8,2.7E9,1E7,-4E8,3E9,1E7))
p.circle(Y[:,0],Y[:,1],color="green")
p.line(Y[:,0],ODMRfit(Y[:,0],*poptY),color='olive')

poptZ, pcov = curve_fit(ODMRfit,Z[:,0],Z[:,1],p0=(-4E8,2.7E9,1E7,-4E8,3E9,1E7))
p.circle(Z[:,0],Z[:,1],color="blue")
p.line(Z[:,0],ODMRfit(Z[:,0],*poptZ),color='navy')
show(p)
```

```python
print "X resonances: %e %e" % (poptX[1],poptX[4])
print "Y resonances: %e %e" % (poptY[1],poptY[4])
print "Z resonances: %e %e" % (poptZ[1],poptZ[4])
```

```python
g = 28E6 #NV gyromagnetic ratio in Hz/mT
Bx = (poptX[4]-poptX[1])/g
By = (poptY[4]-poptY[1])/g
Bz = (poptZ[4]-poptZ[1])/g

print "X field: %.2f mT \nY field: %.2f mT \nZ field: %.2f mT" % (Bx,By,Bz)
```

```python
Bz = Bz/0.88 #scale Bz to account for inhomogenous field strength under z rotation
B = np.sqrt(Bx**2+By**2+Bz**2)
phi = np.arctan(By/Bx)*180/np.pi
theta = np.arccos(Bz/B)*180/np.pi
print "B: %.2f\nPhi: %.2f\nTheta: %.2f" % (B,phi,theta)
```
