```python
t= np.arange(0,1,0.01)
# K_b*T at RT is 4.04*10**-21 J

def spec(f,k,V,t):
    t_n = t*np.exp(k*V/(4.04*10**-21))
    return (2 * t_n)/(np.pi* (1+np.square(t_n*f)))

def F1(f,b,T2):
    w_p = 2.865E9 + 28E6*b
    w_n = 2.865E9 - 28E6*b
    k = 1/T2
    return (1/np.pi)*( k/(k**2 + np.square(f-w_p)) + k/(k**2 + np.square(f-w_n)))
    
def plot_spec(k,r,t_exp,b):
    f = np.logspace(5,10,2000)
    t = 10**t_exp
    V = (4/3.0)*np.pi*(r*10**-9)**3
    T2 = 500E-9
    plt.semilogx(f,spec(f,k,V,t),'k-')
    plt.show()

    
widgets.interact(plot_spec,k=(10**4,4*10**4,10**2),r=(1,10,0.1),t_exp=(-13,-9,0.1),b=(0,100,0.5))
```

```python
def b_var(T2,T1,r):
    b=0
    k =  mp.mpf(26E3)
    t =  mp.mpf(10**-11)
    T1 = mp.mpf(T1*1E-6)
    T2 = mp.mpf(T2*1E-9)
    V =  mp.mpf(4/3.0*np.pi*(r*1E-9)**3)
    i = lambda f: F1(f,b,T2)*spec(f,k,V,t)
    S = mp.quad(i,[10E3,10E11])
    G_ext = mp.mpf(1/mp.mpf(700E-9)  - 1/T1)
    B2 = mp.mpf( G_ext / (mp.mpf(28E6)**2*S))
    return mp.sqrt(B2)

widgets.interact(b_var,T2=(50,2500,0.1),T1=(5,100,0.1),r=(1,10,0.1))
```

```python
T2 = 500
T1 = 15
r = np.arange(1,8,0.05)
b = list( b_var(T2,T1,R) for R in np.nditer(r))
plt.plot(r,b,'k-o')
```

```python
def T1_pred(b,r,b_v):

    
    k =  mp.mpf(26E3)
    t =  mp.mpf(10**-11)
    T1 = mp.mpf(15*1E-6)
    T2 = mp.mpf(500*1E-9)
    V =  mp.mpf(4/3.0*np.pi*(r*1E-9)**3)
    i = lambda f: F1(f,b,T2)*spec(f,k,V,t)
    S = mp.quad(i,[10E3,10E11])
    #return S
    G_ext = mp.mpf(28E6**2*b_v**2*S)
    return mp.mpf(1/(1/T1+G_ext))

def plot_T1(r):
    T2 = 500
    T1 = 15
    b_v = b_var(T2,T1,r)
    b = np.arange(0,50,0.5)
    T1 = list( T1_pred(B,r,b_v) for B in np.nditer(b))
    plt.plot(b,T1,'k-o')
    plt.show()


plot_T1(5)
```
