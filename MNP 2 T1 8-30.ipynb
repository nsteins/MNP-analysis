```python
import glob
def EPR(EPRfiles,Pfiles):
    files = glob.glob(EPRfiles)
    pseq = np.loadtxt(Pfiles)
    d_sum = np.zeros((pseq.shape[0],))
    l = pseq.shape[0]
    for f in files:
        epr = np.loadtxt(f)
        diff = (epr[l:,0]-epr[l:,1])
        diff[np.isinf(diff)] = 0
        diff = np.nan_to_num(diff)
        diff.resize(((diff.shape[0]- (diff.shape[0] % l)),1))
        diff = np.reshape(diff,(pseq.shape[0],-1),order='F')
        d_sum += np.mean(diff,axis=1)
        
    return pseq,d_sum/len(files)

time,T0 = EPR(r'C:\Users\Ouyang Group\Documents\Data\8-27-16\MNP on CPW 5mW scan 2 x54 y120  T1 *[0-9]',
              r'C:\Users\Ouyang Group\Documents\Data\8-27-16\MNP on CPW 5mW scan 2 x54 y120 T1 Pulse Seq.txt')

plt.semilogx(time+1000,T0,'ko:')
print T0
```

```python
def T1Fit(t,I_inf,C_m,T_m,C_1,T_1):
         return I_inf * (1 - C_m*np.exp(-t/T_m) + C_1*np.exp(-t/T_1))
```

```python
import glob
def EPR_Test(EPRfiles,pseq):
    files = glob.glob(EPRfiles)
    d_sum = np.zeros((pseq.shape[0],))
    l = pseq.shape[0]
    for f in files:
        epr = np.loadtxt(f)
        diff = epr[:,0]
        diff = np.nan_to_num(diff)
        diff.resize(((diff.shape[0] - (diff.shape[0] % l)),1))
        diff = np.reshape(diff,(pseq.shape[0],-1),order='F')
        d_sum += np.sum(diff,axis=1)
        
    return d_sum

t = np.arange(500,1500,10)
o = EPR_Test(r'C:\Users\Ouyang Group\Documents\Data\8-30-16\MNP on CPW 0.5mW scan 2 x48 y70  T-23 d3.9 2 0.4V test pulse MW ON 2',t)
plt.plot(t,o,'bo-')

o2 = EPR_Test(r'C:\Users\Ouyang Group\Documents\Data\8-30-16\MNP on CPW 0.5mW scan 2 x48 y70  T-23 d3.9 2 0.4V test pulse MW OFF 2',t)
plt.plot(t,o2,'go-')
```
