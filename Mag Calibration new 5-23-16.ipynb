```python
%matplotlib inline
from matplotlib.widgets import Cursor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook
output_file("MagZ.html")
TOOLS = "box_zoom,box_select,crosshair,hover,resize,reset"
```

```python
Xsweep = np.loadtxt(r'C:\Users\Ouyang Group\Documents\Data\GMW mag calibration\5-23-16\current 6A X sweep.txt')
plt.plot(Xsweep[:,1],Xsweep[:,4:])
```

```python
p1 = figure(tools=TOOLS)
p1.circle(Xsweep[:,1],np.abs(np.arctan(Xsweep[:,6]/Xsweep[:,5])*180/np.pi))
show(p1)
```

```python
YRotsweep = np.loadtxt(r'C:\Users\Ouyang Group\Documents\Data\GMW mag calibration\5-23-16\current 6A Y sweep T90.txt')
plt.plot(YRotsweep[:,2],YRotsweep[:,4:])
```

```python
output_file("MagZRotated.html")
p1 = figure(tools=TOOLS)
p1.circle(YRotsweep[:,2],np.abs(np.arctan(YRotsweep[:,4]/YRotsweep[:,5])*180/np.pi))
show(p1)
```

```python
Tsweep = np.loadtxt(r'C:\Users\Ouyang Group\Documents\Data\GMW mag calibration\5-23-16\current 6A T sweep.txt')
plt.plot(Tsweep[:,0],Tsweep[:,4:]/0.0025)
```

```python
output_file("MagZScaling.html")
p2 = figure(tools=TOOLS)
p2.circle(Xsweep[:,1],np.sqrt(Xsweep[:,6]**2+Xsweep[:,5]**2)/np.sqrt(Xsweep[0,6]**2+Xsweep[0,5]**2))
show(p2)
```
