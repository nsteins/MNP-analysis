```python
%matplotlib inline
from pylab import rcParams
rcParams['figure.figsize'] = 12, 8
from NV import *

def T1Fit(t,I_inf,C_m,T_m,C_1,T_1):
         return I_inf * (1 - C_m*np.exp(-t/T_m) + C_1*np.exp(-t/T_1))
```

```python
time,T0 = EPR(r'C:\Users\Ouyang Group\Documents\Data\12-9-16\Nabond ND on CPW 2mW scan 2 multi T1  x6 y65 T1 7200 sec *[0-9]',
         r'C:\Users\Ouyang Group\Documents\Data\12-9-16\Nabond ND on CPW 2mW scan 2 multi T1  x6 y65 T1 7200 sec  Pulse Seq')
time = time+20
T0 = T0 - np.min(T0)
opt0,cov0 = curve_fit(T1Fit,time[1:],T0[1:],maxfev=15500,p0=(0,0.1,200,0.2,15000))
t = np.logspace(1,6,200)
plt.semilogx(t,T1Fit(t,*opt0),'b-')
print opt0
plt.semilogx(time,T0,'ko:')
```
