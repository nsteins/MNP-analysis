```python
df = pd.read_table(r'C:\Users\Ouyang Group\Documents\Data\8-12-16\wavetek calibration new 2')
o2 = np.empty((np.unique(df.v1).shape[0],5))
for i,v in enumerate(np.unique(df.v1)):
    plt.plot(df[df.v1==v].v2,df[df.v1==v].p,'o:')
    o2[i,0] = v
    o2[i,1] = df[df.v1==v].v2.values[(np.abs(df[df.v1==v].p.values - -4)).argmin()]
    o2[i,2] = df[df.v1==v].v2.values[(np.abs(df[df.v1==v].p.values - -4)).argmin()+1]
    o2[i,3] = df[df.v1==v].p.values[(np.abs(df[df.v1==v].p.values - -4)).argmin()]
    o2[i,4] = df[df.v1==v].p.values[(np.abs(df[df.v1==v].p.values - -4)).argmin()+1]
plt.ylim([-20,7])
```
